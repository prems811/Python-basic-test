```python
# This is my Parent class
class Product:
    def __init__(self, id, name):
        self.id = id
        self.name = name

    def display(self):
        print(f"ID is =  {self.id}, Name is = {self.name}")


# Child class A
class A(Product):
    def __init__(self, id, name, count, category):
        super().__init__(id, name)
        self.count = count
        self.category = category

    def display(self):
        print(f"ID is = {self.id}, Name is = {self.name}, Count is = {self.count}, Category is = {self.category}")


# Child class B
class B(Product):
    def __init__(self, id, name, count, category):
        super().__init__(id, name)
        self.count = count
        self.category = category

    def display(self):
        print(f"ID is = {self.id}, Name is = {self.name}, Count is = {self.count}, Category is = {self.category}")


# Child class C
class C(Product):
    def __init__(self, id, name, count, category):
        super().__init__(id, name)
        self.count = count
        self.category = category

    def display(self):
        print(f"ID is = {self.id}, Name is = {self.name}, Count is = {self.count}, Category is = {self.category}")


# Creating object of product class.
p = Product(78, "Amul")
p.display()

# Creating object of A class.
a = A(78, "Amul", 50, "butter")
a.display()

# Creating object of B class.
b = B(78, "Amul", 90, "Milk")
b.display()

# Creating object of C class.
c = C(78, "Amul", 56, "choco")
c.display()
```
