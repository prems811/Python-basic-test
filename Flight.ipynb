```python
# 1st question.
# This is my Flight class.
class Flight:
    # Now I am initializing the attributes using constructor.
    def __init__(self, flight_no, source, destination, base_fare):
        self.flight_no = flight_no
        self.source = source
        self.destination = destination
        self.base_fare = base_fare

    def get_flight_info(self):
        return f"Flight number = {self.flight_no} , Source =  {self.source}, Destination =  {self.destination}"

    def calculate_fare(self, passenger_count, discount_percent=0):
        total = self.base_fare * passenger_count
        if discount_percent > 0:
            total -= total * (discount_percent / 100)
        return total

    def update_route(self, destination, source=None):
        if source:
            self.source = source
        self.destination = destination


# d) creating flight object here.
f = Flight("AI101", "Delhi", "Mumbai", 5000)

print(f.get_flight_info())

# passenger count
print("Fare (no discount):", f.calculate_fare(3))

# passenger count + discount
print("Fare (with discount):", f.calculate_fare(3, 10))

# update only destination
f.update_route("Bangalore")
print(f.get_flight_info())

# update both source and destination
f.update_route("Chennai", "Hyderabad")
print(f.get_flight_info())
```
